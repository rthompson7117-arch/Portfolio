```python
# loop -> clean data -< check duplicates -> store results
"""
useful for:
*email list cleaning
*csv data cleaning
*log processing
*scraping scripts
*automation jobs

emails = [
    " Ryan@email.com ",
    "SARAH@email.com",
    "mike@email.com ",
    "  Ryan@email.com"
]
seen = set()

for email in emails:
  clean = email.strip().lower()

  if clean not in seen:
    print(clean)
    seen.add(clean)
```
