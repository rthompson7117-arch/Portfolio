```python
import os #imports module of tools to work with your computer's folders

files = os.listdir() #returns a list of filenames in current folder
print(files)

count = 1

for file in files:
  if file.endswith(".jpg"):
    new_name= f"product_{count}.jpg"
    os.rename("oldname.jpg", "newname.jpg")
    print(f"Renamed {file} -> {new_name}")
    count += 1
#changes filename and only renames .jpg files, skipping other file types


"""
Version below sorts the files first so the order is more predictable:

import os

files = sorted(os.listdir())

count = 1

for file in files:
  if file.endswith(".jpg"):
    new_name = f"product {count}.jpg"
    os.rename(file, new_name)
    print(f"Renamed {file} -> {new_name}")
    count += 1
"""
```
