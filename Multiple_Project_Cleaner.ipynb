```python
projects = ["Data Science", "Python automation", "AI Tools"]

clean_projects = []

for project in projects:
    clean = project.lower().replace(" ", "_")
    clean_projects.append(clean)

print(clean_projects)
```

```python
#Instant way to clean lists

projects = ["Data Science", "Python Automation", "AI Tools"]

clean = [p.lower().replace(" ", "_") for p in projects]

print(clean)

Output

['data_science', 'python_automation', 'ai_tools']
```

```python

names = ["Alice", "", "Bob", "", "Charlie"]

clean_names = [name for name in names if name != ""]

print(clean_names)

Output:

['Alice', 'Bob', 'Charlie']

""" This removes blank rows.

People get paid to do this kind of thing with Python. """
```

```python
#Convert product names and remove blanks:

products = ["Blue Shirt", "", "Red Hat", "", "Leather Wallet"]

clean = [p.lower().replace(" ", "_") for p in products if p != ""]

print(clean)

Output:

['blue_shirt', 'red_hat', 'leather_wallet']

""" This type of task appears constantly in:

spreadsheets

CSV data

database imports

product catalogs"""
```
