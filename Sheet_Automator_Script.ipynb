```python
import pandas as pd

# load spreadsheet
df = pd.read_csv("data.csv")

# clean columns
df["name"] = df["name"].str.title()
df["city"] = df["city"].str.title()

# remove duplicates
df = df.drop_duplicates()

# rename columns
df.columns = ["Name", "City", "Email"]

# save cleaned file
df.to_csv("cleaned_data.csv", index=False)

print("File cleaned successfully.")
```
