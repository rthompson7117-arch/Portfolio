```python



file_name = input("Enter filename: ")
word2search = input("Enter word to search for: ")

file = open(file_name)

text = file.read() #creates string that reads the contents of the file

# text.count(word2search) this one is optional

count = text.count(word2search) # creates variable that has the count of what is read in the file

print(f"The word {word2search} appears {count} times in {file_name}")

file.close() # closes the file when finished, remember this step

# to search for words "error" and "Error" use text.lower().count("error")


"""method chaining for "data":
data.strip().lower().replace(" ", "-")

convert to title case:
str.title()   #where just the 1st character of every word is capitalized

"""


```
