```python
"""
Example:

IMG_001.jpg
IMG_002.jpg
IMG_003.jpg

Becomes:

product_1.jpg
product_2.jpg
product_3.jpg
Example Code
"""
import os

folder = "images"

files = os.listdir(folder)

for i, file in enumerate(files):

    new_name = f"product_{i}.jpg"

    os.rename(
        f"{folder}/{file}",
        f"{folder}/{new_name}"
    )
```
