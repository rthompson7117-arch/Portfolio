```python
import os

files = os.listdir()

count = 1

for file in files:
  if file.endswith(".jpg"):
    new_name = f"image_{count}.jpg"
    os.rename(file, new_name)
    print(f"Renamed {file} -> {new_name}")
    count += 1
```
