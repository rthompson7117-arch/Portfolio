```python
"""
Automatic Report Generator

This script creates daily reports automatically.

Example:

Daily Sales Report
Date: 2026-03-05

Total Orders: 24
Revenue: $1,245
Example Code
"""

import datetime

today = datetime.date.today()

report = f"""
Daily Report
Date: {today}
"""

with open("report.txt", "w") as f:
    f.write(report)
```
