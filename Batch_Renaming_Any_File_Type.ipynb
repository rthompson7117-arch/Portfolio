```python
#handles multiple file types
# preserves extensions
# avoids renaming non-images
# works on folders with lots of other file types
# this only works when you're running it in the folder you're cleaning

import os

files = sorted(os.listdir())

count = 1

for file in files:

  name, ext = os.path.splitext("photo.jpg") #splits filename into filename + extension "img_001" and ".jpg"

  if ext.lower() in [".jpg", ".png", ".jpeg"]:

      new_name = f"image_{count}{ext}"

      os.rename(file, new_name)

      print(f"Renamed {file} -> {new_name}")

      count += 1
print(name)
print(ext)


```
